# Simple Linear Regression
y' = b0 + b1X1<br>
Dependent Variable = b0(y-intercept)(constant) + X1(Independent Variable).b1(Slope Coefficient)

## Example
### Predicting Potatoes and seeing the relation to Fertilizers
y' = b0 + b1X1<br>
Potatoes[t] = b0 + b1xFertilizer[kg]<br><br>
So, the model will figure out these values for us like below.<br>
b0 = 8[t], b1 = 3[t/kg]<br><br>


### Ordinary Least Squares
Yi actaual value<br>
yi^ predicted value<br>
residual Ei = yi - y-^<br>
y^ = b0 + b1X1<br><br>
Our goal is to get the b0 and b1 values such that the sum of the residuals is minimized.<br>
ie, the Sum of (yi - yi^)^2 is minimized




What is the constant in simple linear regression?<br>
It is where the line crosses the vertical axes


## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
print(dataset)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
#call the fit method for training the model
#fit method is a method of the LinearRegression class
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```python
print(X_test)
y_pred = regressor.predict(X_test)
print(y_pred)
```

### Check the salary of an employee with x years of experience
We have used [[]] because predict method expects a 2d array as input

```python
print(regressor.predict([[25],[30]]))
```

## Visualising the Training set results

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(X_test, y_test, color = 'blue')
#plt.scatter(X_test, y_pred, color = 'green')
plt.plot(X_test, y_pred, color = 'red')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

### Getting the final linear regression equation with the values of the coefficients

```python
print(regressor.coef_)
print(regressor.intercept_)
```

Therefore, the equation of our simple linear regression model is:

$$\textrm{Salary} = 9345.94 \times \textrm{YearsExperience} + 26816.19$$

**Note:** To get these coefficients we called the "coef_" and "intercept_" attributes from our regressor object
